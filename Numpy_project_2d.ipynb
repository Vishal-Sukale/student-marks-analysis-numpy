## Numpy Project - Students Marks Analysis Using 2D Array

### Objective
 - To analyze student wise and subject wise marks using 2D Numpy array and perform academic Performance analysis.  

### Dataset(2D Array)
- Each row = Students
- Each Column = Subject

Subjects:

1. Machine Learning
2. Deep Learning
3. Math
4. Science


```python
import numpy as np

# Students Marks Data
marks = np.array([
    [78,85,88,90], # Student1
    [65,70,72,75], # Student2
    [92,90,95,94], # Student3
    [55,60,58,62], # Student4
    [80,82,85,88]  # Student5
    ])
```

### Basic Analaysis

```python
print("Total Number of Elements :",marks.size)
print("Shape of Array :", marks.shape)
print("Dimension of Array :", marks.ndim)
print("Type of Data :",type(marks))
print("The Highest Marks :",np.max(marks))
print("The Lowest Marks : ", np.min(marks))
```

- Array Overview : A **2D NumPy** array with **20** values arranged neatly in **5 rows** and **4 columns**.
- Array Type : Stored as a **NumPy ndarray**, enabling fast and efficient data processing.
- Key Results : The **highest** mark is **95** and the **lowest** mark is **55**, showing the data range clearly.


### Analysis & Operations

#### 1. Total Marks Per Student & Per Subject

```python
student_total = np.sum(marks,axis=1)
print("Total Marks Per Student :", student_total)

subject_total = np.sum(marks,axis=0)
print("Total Marks Per Subject :", subject_total)
```

- Student-wise Total : `axis=1` adds marks row-wise to calculate the total marks for each student.
- Subject-wise Total : `axis=0` adds marks column-wise to calculate the total marks for each subject.


#### 2. Average Marks Per Student and Per Subject

```python
student_avg = np.mean(marks,axis=1)
print("Average Marks Per Student :", student_avg)

subject_avg = np.mean(marks,axis=0)
print("Average Marks Per Subject :", subject_avg)
```

- Student-wise Average : `axis=1` calculates the average marks for each student by averaging values across rows.
- Subject-wise Average : `axis=0` calculates the average marks for each subject by averaging values down columns.


#### 3. Result Status
(condition: Average >= 60 - Pass)

```python

Result = np.where(student_avg >= 60, "Pass","Fail")
print("The Result Status Of Students :",Result)
```

- Result Evaluation : Checks each student’s average marks to determine their result status.
- Pass/Fail Logic : Students with an average **≥ 60** are marked **“Pass”**, otherwise **“Fail”**.


#### 4. Performance Grading

```python
grades = np.where(student_avg >= 90, "A",
         np.where(student_avg >= 75, "B",
         np.where(student_avg >= 60, "C", "F")))
print( "The Grade performance of students :",grades)
```

- Grade Assignment : Uses nested `np.where()` to assign grades based on each student’s average marks.
- Grading Scale : **A (≥90)**, **B (≥75)**, **C (≥60)**, and **F (<60)** for clear performance classification.


#### 5. Topper Identification

```python
topper = np.argmax(student_total)
print("The student",topper + 1,"is Topper")
```

- Topper Identification : `np.argmax()` finds the index of the student with the **highest total marks**.
- Student Numbering : `+1` is added to match **human-friendly student numbering** (starting from 1).
